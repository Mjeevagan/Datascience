```python
#to replace multiple columns in the dataset 
#segrate quan and Qual 
def QuanQual(BKC):
    Quan=[]
    Qual=[]
    for ColumnName in BKC.columns:
        if (BKC[ColumnName].dtype=='O'):
            Qual.append(ColumnName)
        else:
            Quan.append(ColumnName)
    return Quan,Qual
```

```python
Quan,Qual=QuanQual(BKC)
```

```python
BKC[Quan]
```

```python
#replace the multiple null values in the dataset using impute method in python
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='median')
imp.fit(BKC[Quan])
BKC=imp.transform(BKC[Quan])
BKC=pd.DataFrame(BKC,columns=Quan)
```
